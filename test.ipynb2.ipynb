```python
##function
def is_strong_password(password):
    # Check length
    if len(password) < 8:
        return False

    # Check for digits
    if not any(char.isdigit() for char in password):
        return False

    # Check for lowercase
    if not any(char.islower() for char in password):
        return False

    # Check for uppercase
    if not any(char.isupper() for char in password):
        return False

    # Check for special characters
    if not any(char in '!@#$%&*()_+' for char in password):
        return False

    return True


## calling the function
print(is_strong_password("hgujjk"))      # False (too short, no digits, no uppercase, no special)
print(is_strong_password("strongpwd!"))  # False (no uppercase, no digits)
print(is_strong_password("Strong1!"))    # True
```

```python
def bubble_sort(arr):
    n = len(arr)
    # Outer loop for passes
    for i in range(n):
        # Inner loop for comparisons
        for passes in range(4):
            for j in range(0, n - i - 1):
                if arr[j] > arr[j + 1]:
                    arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr


# Test the function
unsorted_list = [12, 25, 35, 11, 45]
sorted_list = bubble_sort(unsorted_list)
print("Sorted Elements:", sorted_list)
```

```python
def insertion_sort(lst):
    # Your code goes here
    for i in range(1,len(lst)):
        key=lst[i]
        j=i-1
        while j>=0 and lst[j]>key:
            lst[j+1]=lst[j]
            j -= 1 
        lst[j+1]=key
    return lst

#test the function
print(insertion_sort([15,12,50,42,48,31]))
```

```python
#Count negative numbers in a sorted matrix
def countNegatives(grid):
    def count_negatives_in_row(row):
        left, right = 0, len(row)
        while left < right:
            mid = (left + right) // 2
            if row[mid] < 0:
                right = mid
            else:
                left = mid + 1
        return len(row) - left

    total_negatives = 0
    for row in grid:
        total_negatives += count_negatives_in_row(row)
    return total_negatives

print(countNegatives([
    [4, 3, 2, 1],
    [3, 2, 1, -1],
    [1, 1, -1, -2],
    [-1, -1, -2, -3]
]))
```

```python
#Find smallest letter greater than target
def nextGreatestLetter(letters, target):
    left, right = 0, len(letters) - 1
    while left <= right:
        mid = (left + right) // 2
        if letters[mid] <= target:
            left = mid + 1
        else:
            right = mid - 1
    # wrap around
    return letters[left % len(letters)]
#  Test the function
letters = ['c', 'f', 'j']

print(nextGreatestLetter(letters, 'a'))  # Expected: 'c'
print(nextGreatestLetter(letters, 'c'))  # Expected: 'f'
print(nextGreatestLetter(letters, 'd'))  # Expected: 'f'
print(nextGreatestLetter(letters, 'g'))  # Expected: 'j'
print(nextGreatestLetter(letters, 'j'))  # Expected: 'c' (wrap around)
print(nextGreatestLetter(letters, 'k'))  #
```

```python
#Find First and Last Position of Element in Sorted Array
def searchRange(nums, target):
    # Implement your solution here
    def find_first(nums,target):
        left,right=0,len(nums)-1
        first=-1
        while left <= right:
            mid = (left+right)//2
            if nums[mid]==target:
                first=mid
                right=mid-1
            elif nums[mid]<target:
                left=mid+1 
            else:
                right=mid-1 
        return first
    def find_last(nums, target):
        left, right = 0, len(nums) - 1
        last = -1
        while left <= right:
            mid = (left + right) // 2
            if nums[mid] == target:
                last = mid
                left = mid + 1  # keep searching right side
            elif nums[mid] < target:
                left = mid + 1
            else:
                right = mid - 1
        return last

    return [find_first(nums, target), find_last(nums, target)]
# test the function
print(searchRange([5,7,7,8,8,10], 8))   # Output: [3, 4]
print(searchRange([5,7,7,8,8,10], 6))   # Output: [-1, -1]
print(searchRange([], 0))    
```

```python
#rotate the list
def rotate_left(ARR, D):
    n=len(ARR)
    D=D%n

    for _ in range(D):
        first = ARR[0]  # store the first element
        for i in range(n - 1):
            ARR[i] = ARR[i + 1]  # shift elements left
        ARR[-1] = first  # put the first element at the end
    return ARR
#test the function
numbers = [1, 2, 3, 4, 5]
print(rotate_left(numbers, 2))   

    
```

```python
#Plus One in the Number
def plus_one(digits):
    
    n = len(digits)
    
    # Traverse from the last digit backward
    for i in range(n - 1, -1, -1):
        if digits[i] < 9:
            digits[i] += 1
            return digits  # No carry, return immediately
        digits[i] = 0  # Set current digit to 0 if it becomes 10

    # If all digits were 9 (like [9,9,9]), add 1 at the start
    return [1] + digits
#Example Test Cases

print(plus_one([1, 2, 3]))   # Output: [1, 2, 4]
print(plus_one([4, 3, 9]))   # Output: [4, 4, 0]
print(plus_one([9, 9, 9]))   # Output: [1, 0, 0, 0]
```

```python
#find_missing_number
def find_missing_number(nums):
    n = len(nums)
    expected_sum = n * (n + 1) // 2   # Sum of numbers from 0 to n
    actual_sum = sum(nums)            # Sum of all given numbers
    return expected_sum - actual_sum  
#test the function 
print(find_missing_number([3, 0, 1]))   # Output: 2
print(find_missing_number([0, 1]))      # Output: 2
print(find_missing_number([9,6,4,2,3,5,7,0,1]))
```

```python
def find_max_consecutive_ones(nums):
    """
    Function to find the maximum number of consecutive 1's in a binary array.
    """
    max_count = 0     #      1s  streak
    current_count = 0 #    1s  streak
    
    for num in nums:
        if num == 1:
            current_count += 1
            max_count = max(max_count, current_count)
        else:
            current_count = 0  # 0   streak 
    
    return max_count
#test the function 
print(find_max_consecutive_ones([1, 1, 0, 1, 1, 1]))  #  3
print(find_max_consecutive_ones([1, 0, 1, 0, 1]))     #  1
print(find_max_consecutive_ones([0, 0, 0]))           #  0
print(find_max_consecutive_ones([1, 1, 1, 1]))        
```

```python
def max_subarray_sum(arr):
    """
    Given an array of integers, find the maximum sum of any subarray using Kadane's Algorithm.
 
    Parameters:
    arr (List[int]): List of integers.
 
    Returns:
    int: Maximum sum of any subarray.
    """
    if not arr:
        return 0
 
    # Initialize variables
    max_ending_here = 0
    max_so_far = float('-inf')
 
    # Iterate through the array
    for x in arr:
        # Update max_ending_here to include current element
        max_ending_here = max(x, max_ending_here + x)
        # Update max_so_far to the maximum value found so far
        max_so_far = max(max_so_far, max_ending_here)
 
    return max_so_far
```

Python OOPs (Object-Oriented Programming System)
 What is OOPs?

OOPs (Object-Oriented Programming) is a programming approach based on real-world entities like class, object, inheritance, polymorphism, abstraction, and encapsulation.

1. Class

A class is a blueprint or template for creating objects.

```python
class Car:
    def __init__(self, brand, color):
        self.brand = brand
        self.color = color

    def show(self):
        print(f"{self.brand} - {self.color}")
```

2. Object

An object is an instance of a class.

```python
c1 = Car("BMW", "Black")
c1.show()   # Output: BMW - Black
```

3. Inheritance

Inheritance allows a class to use properties and methods of another class.

```python
class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

d = Dog()
d.speak()  # Animal speaks
d.bark()   # Dog barks
```

Types of Inheritance:

Single  One parent, one child

Multiple  Multiple parents, one child

Multilevel  Parent  Child  Grandchild

Hierarchical  One parent, many children

Hybrid  Combination of above

Polymorphism

Same function name, but different behaviors

```python
class Bird:
    def intro(self):
        print("There are many birds")

class Sparrow(Bird):
    def intro(self):
        print("Sparrow is small")

b = Sparrow()
b.intro()  # Sparrow is small
```

Encapsulation

Hiding data using private or protected members.

```python
class Account:
    def __init__(self, balance):
        self.__balance = balance  # private variable

    def show_balance(self):
        print(self.__balance)
```

6. Abstraction

Hiding internal implementation and showing only the necessary functionality.

```python
from abc import ABC, abstractmethod

class Car(ABC):
    @abstractmethod
    def mileage(self):
        pass

class BMW(Car):
    def mileage(self):
        print("Mileage is 25km/l")

b = BMW()
b.mileage()
```

Advanced Python Topics

 1. Iterator

An iterator is an object used to iterate over a sequence using next().

```python
class Count:
    def __init__(self, start, end):
        self.num = start
        self.end = end

    def __iter__(self):
        return self

    def __next__(self):
        if self.num <= self.end:
            val = self.num
            self.num += 1
            return val
        else:
            raise StopIteration

for i in Count(1, 5):
    print(i)
```

2. Generator

A generator is a special type of function that yields values one at a time using the yield keyword.

```python
def squares(n):
    for i in range(1, n+1):
        yield i * i

for val in squares(5):
    print(val)
```

Advantages:

Memory efficient

Lazy evaluation (values are generated only when needed)

3. Decorator

A decorator is a function that modifies another functions behavior without changing its code.

```python
def decorator_func(func):
    def wrapper():
        print("Before function call")
        func()
        print("After function call")
    return wrapper

@decorator_func
def say_hello():
    print("Hello World!")

say_hello()
```

4. Lambda Function

Small anonymous (unnamed) function using lambda keyword

```python
add = lambda a, b: a + b
print(add(2, 3))
```

Sum of Natural Numbers
By Recursion
Finding the sum of natural numbers is a simple and classic example used to illustrate recursion.
Problem: Find the sum of all natural numbers up to n.
Example: For n=5, the sum is 1+2+3+4+5 = 15

Definition
Let S(n) be the sum of natural numbers up to n.We can write it as:
S(n) = 1 + 2 + 3 +..... + (n-1) + n

Recursive Relation
By observing the equation above, we can identify that the sum from 1 to (n-1) is equal to S(n-1).
S(n) = [1 + 2 + 3 +....+ (n-1)] + n
S(n) = S(n-1) + n
This is the Recursive Relation, which defines S(n) in terms of S(n-1) for any number n.


Base Case
Every recursive function requires a Base Case that stops the recursion at a definite point.
If n=1, the sum is 1.
S(1) = 1
When $n$ reaches 1, the function starts returning, and the recursion terminates.

```python
def factorial(n):
    """
    Function to calculate the factorial of a non-negative integer n using recursion.
    
    Parameters:
    n (int): The non-negative integer for which the factorial is to be calculated.
    
    Returns:
    int: The factorial of n.
    """
    if n == 0 or n == 1:
        return 1
    # Recursive case: n * factorial of (n-1)
    else:
        return n * factorial(n - 1)
```

```python
def sum_of_natural_numbers(n):
    """
    Function to calculate the sum of the first n natural numbers using recursion.
    
    Parameters:
    n (int): The non-negative integer for which the sum is to be calculated.
    
    Returns:
    int: The sum of the first n natural numbers.
    """
    if n == 0:
        return 0
    # Recursive case: n + sum of numbers up to (n-1)
    else:
        return n + sum_of_natural_numbers(n - 1)
```

```python
##First Index of an Element using Recursion
def find_first_index(arr, element):
    """
    Function to find the first index of a given element in an array using recursion.
    
    Parameters:
    arr (list of int): The array to search through.
    element (int): The element to find.
    
    Returns:
    int: The first index of the element in the array, or -1 if not found.
    """

    #  Base case: if array is empty, element not found
    if len(arr) == 0:
        return -1

    #  If the first element matches, return 0 (index of first element)
    if arr[0] == element:
        return 0

    #  Otherwise, check in the remaining array (recursive call)
    index_in_rest = find_first_index(arr[1:], element)

    #  If not found in rest, return -1
    if index_in_rest == -1:
        return -1
    else:
        #  Otherwise, add +1 to adjust for the skipped first element
        return index_in_rest + 1
    #test the function
print(find_first_index([3, 2, 5, 2, 8, 2, 1], 3))  # 0
print(find_first_index([3, 2, 5, 2, 8, 2, 1], 5))  # 2
print(find_first_index([3, 2, 5, 2, 8, 2, 1], 1))  # 6
print(find_first_index([3, 2, 5, 2, 8, 2, 1], 9))  # -1

```

```python
def find_indices(arr, element):
    """
    Function to find all indices of a given element in an array using recursion.
    
    Parameters:
    arr (list of int): The array to search through.
    element (int): The element to find.
    
    Returns:
    list of int: A list containing all indices of the element in the array.
    """
     # Helper function to perform the recursive search
    def helper(index):
        # Base case: if index is equal to the length of the array, return empty list
        if index == len(arr):
            return []
        
        # Check if the current index contains the element
        if arr[index] == element:
            # Include the current index and proceed with the next index
            return [index] + helper(index + 1)
        else:
            # Proceed with the next index without including the current index
            return helper(index + 1)
 
    # Start the recursive search from index 0
    return helper(0)
# Example calls 
print(find_indices([3, 2, 5, 2, 8, 2, 1], 2))
print(find_indices([1, 2, 3, 4, 2, 5, 2], 2))
print(find_indices([3, 2, 5, 2, 8, 2, 1], 8))
print(find_indices([3, 2, 5, 2, 8, 2, 1], 9))

```

```python
def linear_search(arr, target):
    """
    Function to perform linear search on an array using recursion.
    
    Parameters:
    arr (list of int): The array of integers.
    target (int): The element to search for.
    
    Returns:
    bool: True if target is found, False otherwise.
    """
    if not arr:
        return False
    
    # Base Case 2: The current element (first element) is the target.
    if arr[0] == target:
        return True
    
    # Recursive Step: Check the rest of the array (from the second element onwards).
    # This involves slicing the array, which creates a new list in each call.
    return linear_search(arr[1:], target)

# --- Alternative implementation passing index/start point to avoid slicing ---

def linear_search_alt(arr, target, index=0):
    """
    Alternative recursive linear search using an index/start point,
    which is generally more memory efficient by avoiding list slicing.
    
    Parameters:
    arr (list of int): The array of integers.
    target (int): The element to search for.
    index (int): The current starting index to check from (default is 0).
    
    Returns:
    bool: True if target is found, False otherwise.
    """
    # Base Case 1: The index has gone past the end of the array. Target is not found.
    if index >= len(arr):
        return False
    
    # Base Case 2: The element at the current index is the target.
    if arr[index] == target:
        return True
    
    # Recursive Step: Check the rest of the array by incrementing the index.
    return linear_search_alt(arr, target, index + 1)
```

```python
# Function to perform binary search recursively
def BinarySearchUsingRecursionHelper(l1, x, s, e):
    # Base Case 1: Search space is exhausted (start > end). Target not found.
    if s > e:
        return False

    # Calculate the middle index 'm'. 
    # Use s + (e - s) // 2 to prevent potential integer overflow for very large s and e,
    # though standard (s + e) // 2 is usually fine in Python.
    m = s + (e - s) // 2

    # Base Case 2: Element at the middle index 'm' is the target. Target found.
    if l1[m] == x:
        return True

    # Recursive Step 1: If the target 'x' is greater than the middle element,
    # search the right half of the array (from m + 1 to e).
    if x > l1[m]:
        return BinarySearchUsingRecursionHelper(l1, x, m + 1, e)

    # Recursive Step 2: If the target 'x' is less than the middle element,
    # search the left half of the array (from s to m - 1).
    else:  # x < l1[m]
        return BinarySearchUsingRecursionHelper(l1, x, s, m - 1)


# Wrapper function to start the binary search
def BinarySearchUsingRecursion(l1, x):
    # Call the helper function, starting the search from index 0 to the last index (len(l1) - 1).
    return BinarySearchUsingRecursionHelper(l1, x, 0, len(l1) - 1)


# --- Example Usage from the screenshot ---

# Create a sorted list l1 containing numbers from 0 to 999,999
# Note: Binary search requires the list to be sorted.
l1 = []
for i in range(1000000):
    l1.append(i)

# Search for the element 1200 in the list l1
print(BinarySearchUsingRecursion(l1, 1200))
```

```python
def merge_sort(arr):
    """
    Function to perform merge sort on a list of integers using recursion.
    
    Parameters:
    arr (list of int): The list to be sorted.
    
    Returns:
    list of int: The sorted list.
    """
    # Base case: if list has 0 or 1 element, it is already sorted
    if len(arr) <= 1:
        return arr
    
    # Step 1: Divide the array into two halves
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    
    # Step 2: Merge the two sorted halves
    return merge(left, right)


def merge(left, right):
    sorted_list = []
    i = j = 0
    
    # Compare and merge
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            sorted_list.append(left[i])
            i += 1
        else:
            sorted_list.append(right[j])
            j += 1
    
    # Add remaining elements
    sorted_list.extend(left[i:])
    sorted_list.extend(right[j:])
    
    return sorted_list


# Example:
print(merge_sort([5, 2, 4, 6, 1, 3]))
```

```python
def quick_sort(arr):
    """
    Function to perform quick sort on a list of integers using recursion.
    
    Parameters:
    arr (list of int): The list to be sorted.
    
    Returns:
    list of int: The sorted list.
    """
    # 1. Base Case: An array with 0 or 1 element is already sorted.
    if len(arr) <= 1:
        return arr

    # 2. Choose the Pivot: Selecting the last element as the pivot is common.
    # Other common choices are the first element or a random element.
    pivot = arr[-1]
    
    # Initialize lists for elements smaller, equal, and greater than the pivot.
    less = []
    equal = []
    greater = []
    
    # 3. Partitioning: Iterate through the array and place elements 
    # into the appropriate list based on the pivot.
    for x in arr:
        if x < pivot:
            less.append(x)
        elif x == pivot:
            equal.append(x)
        else:
            greater.append(x)
            
    # 4. Recursive Call and Combine:
    # Recursively sort the 'less' and 'greater' lists, and then combine 
    # them with the 'equal' list (where the pivot and its duplicates are).
    # The combination automatically forms the sorted list.
    return quick_sort(less) + equal + quick_sort(greater)

# --- Example Usage ---
my_list = [10, 7, 8, 9, 1, 5]
sorted_list = quick_sort(my_list)

print(f"Original List: {my_list}")
print(f"Sorted List (Quick Sort): {sorted_list}")
```

```python
def remove_character(s1,ch):
    """
    Function to remove all occurrences of a given character from a string using recursion.
    
    Parameters:
    s1 (str): The input string.
    ch (str): The character to be removed.
    
    Returns:
    str: The string after removing all occurrences of the character.
    """
    # Base case: if the string is empty, return an empty string
    if len(s1) == 0 or s1=="":
        return s1
    
    # Recursive case
    if s1[0] == ch:
        # If the first character matches 'ch', skip it and recurse on the rest
        return remove_character(s1[1:], ch)
    else:
        # If it doesn't match, include it and recurse on the rest
        return s1[0] + remove_character(s1[1:], ch)
    
    s1="abcaad"
    ans=remove_character(s1,'d')
    print(s1)
```

```python
def palindrome_helper(s1, start, end):

    if start >= end:
        return True

    if s1[start] != s1[end]:
        return False

    return palindrome_helper(s1, start + 1, end - 1)


def palindrome(s1):
    return palindrome_helper(s1, 0, len(s1) - 1)


print(palindrome('nitin'))
```

```python
##Palindrome Check using Recursion

def is_palindrome(s):
    """
    Function to check if a string is a palindrome using recursion.
    
    Parameters:
    s (str): The string to check.
    
    Returns:
    bool: True if the string is a palindrome, False otherwise.
    """
    if len(s) <= 1:
        return True
    
    # If first and last characters mismatch  not palindrome
    if s[0] != s[-1]:
        return False
    
    # Recursive call on the substring without first and last characters
    return is_palindrome(s[1:-1])
```

```python
def subsequences(s):
    """
    Function to generate all subsequences of a string using recursion.
    
    Parameters:
    s (str): The string to generate subsequences from.
    
    Returns:
    list of str: A list containing all subsequences of the string.
    """

    # Base case: empty string  only one subsequence: ""
    if s == "":
        return [""]

    # Recursive call on the remaining string
    small_ans = subsequences(s[1:])

    # For each subsequence, add the first character also
    result = []
    for sub in small_ans:
        result.append(sub)          # without first character
        result.append(s[0] + sub)   # with first character

    return result
```

```python
def permutations(s):
    """
    Function to generate all permutations of a string using recursion.
    
    Parameters:
    s (str): The string to generate permutations from.
    
    Returns:
    list of str: A list containing all permutations of the string.
    """
    if len(s) == 0:
        return [""]

    # Recursive case
    result = []
    for i in range(len(s)):
        # Pick one character
        ch = s[i]
        # Remaining string
        rest = s[:i] + s[i+1:]
        # Generate permutations of the remaining string
        for perm in permutations(rest):
            result.append(ch + perm)

    return result
```

```python
def get_char(value):
    if value <= 0 or value > 26:
        return ''
    return chr(96 + value)


def return_all_codes(input):
    if input == '':
        return ['']

    singleDigit = int(input[0:1])
    doubleDigit = int(input[0:2])
    mainAns = []

    ansWithoutFirstDigit = return_all_codes(input[1:])

    for eachAns in ansWithoutFirstDigit:
        mainAns.append(get_char(singleDigit) + eachAns)

    if(doubleDigit >= 10 and doubleDigit <= 26):
        ansWithoutDoubleDigit = return_all_codes(input[2:])
        for eachAns in ansWithoutDoubleDigit:
            mainAns.append(get_char(doubleDigit) + eachAns)

    return mainAns


ans = return_all_codes("1123")
print(ans)
```

```python
def tower_of_hanoi(n, source, helper, destination, moves=None):
    """
    Recursive Tower of Hanoi.
    Moves disks from `source` to `destination` using `helper`.
    If moves list is provided, appends each move as a tuple (disk, from, to).
    Returns the total number of moves.
    """
    if moves is None:
        moves = []
    if n == 0:
        return 0
    if n == 1:
        moves.append((1, source, destination))
        print(f"Move disk 1 from {source} to {destination}")
        return 1
    # Move n-1 disks from source to helper
    count1 = tower_of_hanoi(n-1, source, destination, helper, moves)
    # Move nth disk from source to destination
    moves.append((n, source, destination))
    print(f"Move disk {n} from {source} to {destination}")
    # Move n-1 disks from helper to destination
    count2 = tower_of_hanoi(n-1, helper, source, destination, moves)
    return count1 + 1 + count2

if __name__ == '__main__':
    n = 3
    print(f"\nTower of Hanoi moves for n={n}:\n")
    total_moves = tower_of_hanoi(n, 'A', 'B', 'C')
    print(f"\nTotal moves: {total_moves}\n")
```

```python
import numpy as np
import time

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

def measure_time_bubble_sort():
    input_sizes = [10, 100, 1000, 5000, 10000]  # safe sizes

    for size in input_sizes:
        arr = np.arange(size, 0, -1)  # reverse sorted

        start_time = time.time_ns()
        bubble_sort(arr)
        end_time = time.time_ns()

        print(f"Time taken to sort array of size {size}: {end_time - start_time} nanoseconds")

measure_time_bubble_sort()

```
